```python
#Importamos las librerias pandas, numpy y matplotlib respectivamente
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Carga desde un archivo .xlsx sin indice
california= pd.read_csv('California_EUA.csv') 
```

```python
#Verificamos información del DataFrame
california.info()
```

```python
#Imprimo los primeros 5 registro del dataframe
california.head()
```

## **DATOS NULOS**

```python
#Corroboramos valores nulos
valores_nulos=california.isnull().sum().sum()
valores_nulos
```

```python
#Separamos nuestro data frame en variables cualitativas y no cualitativas
data1 = pd.DataFrame(california)
y = data1.select_dtypes(include=[np.float, np.int])
x = data1.select_dtypes(include=[np.object])
```

```python
# primero rellenamos nuestros valores numericos con el promedio y los restantes con 0, esto es por que hay columnas que no tienen ni un solo registro y otras son scores que los airbns no tienen  
y= y.fillna(round(y.mean(),1))
nulosy = y.columns[y.isnull().any()]
y[nulosy] = y[nulosy].fillna('')
valores_nulosy=y.isnull().sum().sum()
valores_nulosy
```

```python
# los registros nulos que son cualitativos se rellenan con 'No registro' para indicar que no se proporciono esta informacion
nulosx = x.columns[x.isnull().any()]
x[nulosx] = x[nulosx].fillna('No registro')
valores_nulosx=x.isnull().sum().sum()
valores_nulosx
```

## **VALORES ATIPICOS: PROCEDIMIENTO "CUANTILES"**

```python
#Método aplicando Cuartiles. Encuentro cuartiles 0.25 y 0.75
percentile25=y.quantile(0.25) #Q1
percentile75=y.quantile(0.75) #Q3
iqr= percentile75 - percentile25

Limite_Superior_iqr= percentile75 + 1.5*iqr
Limite_Inferior_iqr= percentile25 - 1.5*iqr
print("Limite superior permitido \n", Limite_Superior_iqr)
print("Limite inferior permitido \n", Limite_Inferior_iqr)
```

```python
outliers_iqr= y[(y>Limite_Superior_iqr)|(y<Limite_Inferior_iqr)]
outliers_iqr
```

```python
data3_iqr= y[(y<=Limite_Superior_iqr)&(y>=Limite_Inferior_iqr)]
data3_iqr
```

```python
valores_nulos=data3_iqr.isnull().sum()
valores_nulos
```

```python
data4_iqr=data3_iqr.copy()
data4_iqr=data4_iqr.fillna(round(data3_iqr.mean(),1))
data4_iqr[['neighbourhood_group_cleansed', 'bathrooms', 'calendar_updated']] = data4_iqr[['neighbourhood_group_cleansed', 'bathrooms', 'calendar_updated']].fillna('No registro')
data4_iqr.isnull().sum()
```

```python
data_clean = pd.concat([x, data4_iqr], axis=1)
data_clean
```

```python
valores_nulos=data_clean.isnull().sum()
valores_nulos
```

```python
filtro1C= data_clean[data_clean['host_acceptance_rate'] > '50%']
#filtro1C['host_acceptance_rate'].min()
filtro1C['host_acceptance_rate']
```

Sólo los registros con categoría “superhost”

```python
filtro2C= data_clean[data_clean['host_is_superhost'] == 't']
filtro2C['host_is_superhost']
```

```python
data_clean['bathrooms']
filtro5C= data_clean[data_clean['bathrooms'] > '1' ]
filtro5C['bathrooms'].unique()
```

```python
filtro6C= data_clean[(data_clean['price'] > 10000) & (data_clean['property_type'] > 'Entire home')]
filtro6C['price']
```
